```python
# Columnas que seran eliminadas
columnas_a_eliminar = ['Unnamed: 0']

# Se eliminan las columnas, por ser insignificantes para el analisis
df_autosElectricos_EDA = df_autosElectricos_EDA.drop(columnas_a_eliminar, axis=1)
```

```python
df_autosElectricos_EDA.info()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df_autosElectricos_EDA , x='PriceDollar', bins=20, kde=True)
plt.title('Distribución de Precios de Autos Eléctricos')
plt.xlabel('Precio (Dólares)')
plt.ylabel('Frecuencia')
plt.show()
```

**Distribución de Precios de Autos Eléctricos:**
   - La distribución de precios de los autos eléctricos parece estar sesgada hacia la derecha, lo que sugiere que la mayoría de los autos eléctricos en el conjunto de datos tienen precios más bajos.
   - La presencia de la curva de densidad (kde) muestra una tendencia general de los precios de los autos eléctricos en el conjunto de datos.

```python
plt.figure(figsize=(5, 5))
sns.scatterplot(data=df_autosElectricos_EDA , x='Range_Miles', y='Efficiency_Whmi')
plt.title('Rango vs. Eficiencia de Autos Eléctricos')
plt.xlabel('Rango (Millas)')
plt.ylabel('Eficiencia (Wh/mi)')
plt.show()
```

**Rango vs. Eficiencia de Autos Eléctricos:**
   - No parece haber una correlación clara entre el rango y la eficiencia de los autos eléctricos. Esto sugiere que la eficiencia no necesariamente se correlaciona directamente con el rango que puede ofrecer un automóvil eléctrico.


```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df_autosElectricos_EDA, x='Brand', order=df_autosElectricos_EDA['Brand'].value_counts().index)
plt.title('Frecuencia de Autos Eléctricos por Marca')
plt.xlabel('Marca')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45, ha='right')
plt.show()

```

**Frecuencia de Autos Eléctricos por Marca:**
   - La marca más frecuente de autos eléctricos en el conjunto de datos se muestra en el gráfico de barras.
   - La rotación de las etiquetas en el eje x facilita la visualización de todas las marcas presentes en el conjunto de datos.
