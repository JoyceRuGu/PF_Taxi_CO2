```python
# Eliminamos las columnas que no son relevantes
df_sonidos = df_sonidos.drop(eliminar, axis=1)
```

```python
# Mantenemos solo las filas donde 'year' es 2019 (registros más actuales)
df_sonidos = df_sonidos[df_sonidos['year'] == 2019]
```

```python
# Buscamos filas repetidas
df_sonidos.duplicated().sum()

# Eliminamos las filas repetidas
df_sonidos = df_sonidos.drop_duplicates()
```

```python
df_sonidos.head()
```

```python
# Creamos una función que modifica los valores de la columna 'Engine Sound' haciendo que '-1' sea Bajo, '0' sea Medio y '1' sea Alto
def engine_sound(engine):
    if engine == -1:
        return 'Low'
    elif engine == 0:
        return 'Medium'
    elif engine == 1:
        return 'High'
    else:
        return 'Unknown'
```

```python
# Aplicamos la función `engine_sound` para crear la columna 'Engine Sound'
df_sonidos['engine_sound'] = df_sonidos['1_engine_presence'].apply(engine_sound)
```

```python
# Eliminamos la columna '1_engine_presence' ya que no es necesaria al conocer el nivel de sonido del motor
df_sonidos =df_sonidos.drop('1_engine_presence', axis=1)
df_sonidos.head()
```

```python
# Renombramos las columnas y reseteamos el índice
df_sonidos.rename(columns={'borough':'Borough','latitude':'Latitude','longitude':'Longitude','year':'Year',
                              'day':'Day','hour':'Hour','engine_sound':'Engine Sound'}, inplace=True)

df_sonidos.reset_index(drop=True)
```

- Borough: El código del distrito o municipio al que pertenece la ubicación registrada:
1 (Manhattan), 2 (Brooklyn), 3 (Queens), 4 (Bronx) y 5 (Staten Island)
- Latitude: La latitud geográfica de la ubicación.

- Longitude: La longitud geográfica de la ubicación.

- Year: El año en que se registró la observación.

- Day: El día del mes en que se registró la observación.

- Hour: La hora del día en que se registró la observación.

- Engine Sound: Representa el nivel de sonido del motor, categorizado como "Low" (bajo), "Medium" (medio) o "High" (alto).

### Guardar el conjunto de datos limpio

```python
df_sonidos.to_csv('sonidos_cleaned.csv', index=False)
```

### Limpieza y preprocesamiento del conjunto de datos: df_aire

```python
# Seleccionamos las columnas que no son relevantes
eliminar = ['Unique ID', 'Indicator ID', 'Measure', 'Geo Type Name', 'Geo Join ID', 'Start_Date']
```

```python
# Eliminamos las columnas que no son relevantes
df_aire = df_aire.drop(eliminar, axis=1)
```

```python
df_aire.head()
```

```python
# Mantenemos solo las filas donde 'year' es 2021 (registros más actuales)
df_aire = df_aire[df_aire['Time Period'].str.contains('2021')]
```

```python
# Cambiamos los valores de la columna 'Year'
df_aire['Year'] = 2021
```

```python
# Eliminamos duplicados basados en todas las columnas
df_aire = df_aire.drop_duplicates()
```

```python
# Renombramos las columnas y reseteamos el índice
df_aire.rename(columns={'Name': 'Pollutant', 'Time Period':'Year' }, inplace=True)

# Actualizamos el índice
df_aire = df_aire.reset_index(drop=True)
```

- Pollutant: El tipo de contaminante medido en la observación.

- Measure Info: Información adicional sobre la medida del contaminante. En este caso, indica la unidad de medida utilizada para expresar la concentración del contaminante.

- Geo Place Name: El nombre del lugar geográfico donde se realizó la medición. Estos podrían ser nombres de vecindarios, distritos o áreas geográficas específicas donde se recopilaron los datos de contaminación.

- Year: El año en que se realizó la medición de la contaminación.

- Data Value: El valor numérico de la concentración del contaminante medido, expresado en la unidad especificada en la columna "Measure Info".
